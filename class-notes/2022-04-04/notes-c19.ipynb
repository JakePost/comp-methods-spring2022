Example 10.4 The Traveling Salesperson

As an example of simulating annealing, lets look at the traveling salesperson problem.

The problem is one where we have a salesperson who needs to visit N cities on their trip, and we want to figure out what is the shortest route between the N cities.

<br/>

$D = \sum^{N-1}_{i=0} |r_{i+1} - r_i|$



```python
## Exercise 8.1

import numpy as np
import matplotlib.pyplot as plt

h = 0.01
RC = 0.1 # 0.1, 1
time = np.arange(0, 10, h)


def v_input(t):
    return 1 if np.floor(2 * t) % 2 == 0 else -1

v_out = np.zeros(len(time))

for n in range(0, len(time) - 1):
    v_out[n + 1] = v_out[n] + h * (1 / RC * (v_input(time[n]) - v_out[n]))

plt.plot(time, [v_input(t) for t in time])
plt.plot(time, v_out)
plt.plot()
```
