### Relaxation method for two or more variables

If instead we have equations for two variables, the relaxation method works just as well as long as each equation can be written in a form that gives back the variable.

THis si one of the strengths of the relaxation method, it is easily extended to many nonlinear equations of multiple variables

## Bisection

Binary Search, or the bisection method, is an alternative and more robust method for solving nonlinear equations for a single variable x.

With this method if a solution exists in a specified interval, this method will always find it.

You can always rearrange an equation in the form f(x) = 0


```python
# Exercise 6.15
import numpy as np
import matplotlib.pyplot as plt


def p(x):
    return (924 * x ** 6) - (2772 * x ** 5 ) + (3150 * x ** 4) - (1680 * x ** 3) + (420 * x ** 2) - (42 * x) + 1

def p_rime(x):
    return (924 * 6 * x ** 5) - (2772 * 5 * x ** 4) + (3150 * 4 * x ** 3) - (1680 * 3 * x ** 2) + (420 * 2 * x) - 42 + 1

x_plt = np.linspace(0, 1, 100)
y_plt = p(x_plt)

y_p_plt = p_rime(x_plt)

plt.plot(x_plt, y_plt)
plt.plot(x_plt, np.zeros(x_plt.shape))
plt.plot(x_plt, y_p_plt)
plt.ylim(-10, 10)
plt.show()
```

```python
def derivative(x, f):
    return (f(x + 10E-5 / 2) - f(x - 10E-5 / 2)) / 10E-5


def newtons_method(x_in):
    x = x_in
    y = p(x)

    for _ in range(0, 100):
        x = x - p(x) / p_rime(x)
        y_new = p(x)
        if y_new >= y:
            break

        y = y_new

    return x, y
```

```python
x_root = np.zeros(6)
y_root = np.zeros(6)

x_root[0], y_root[0] = newtons_method(0.02)
x_root[1], y_root[1] = newtons_method(0.2)
x_root[2], y_root[2] = newtons_method(0.38)
x_root[3], y_root[3] = newtons_method(0.64)
x_root[4], y_root[4] = newtons_method(0.83)
x_root[5], y_root[5] = newtons_method(0.97)

plt.plot(x_plt, y_plt)
plt.plot(x_plt, np.zeros(x_plt.shape))
plt.scatter(x_root, y_root)
plt.show()
```
